```python
data2018["Average Patient Payment Amount"] = data2018["Total Medicare Payment Amount"] / data2018["Number of Medicare Beneficiaries"]
data2017["Average Patient Payment Amount"] = data2017["Total Medicare Payment Amount"] / data2017["Number of Medicare Beneficiaries"]
data2016["Average Patient Payment Amount"] = data2016["Total Medicare Payment Amount"] / data2016["Number of Medicare Beneficiaries"]
data2015["Average Patient Payment Amount"] = data2015["Total Medicare Payment Amount"] / data2015["Number of Medicare Beneficiaries"]
data2014["Average Patient Payment Amount"] = data2014["Total Medicare Payment Amount"] / data2014["Number of Medicare Beneficiaries"]
data2013["Average Patient Payment Amount"] = data2013["Total Medicare Payment Amount"] / data2013["Number of Unique Beneficiaries"]
data2012["Average Patient Payment Amount"] = data2012["Total Medicare Payment Amount"] / data2012["Number of Unique Beneficiaries"]
data2012.rename(columns = {"NPI": "National Provider Identifier"}, inplace = True)
data2013.rename(columns = {"NPI": "National Provider Identifier"}, inplace = True)
```

```python
data2018 = data2018[["National Provider Identifier", "Average Patient Payment Amount"]]
data2018["Year"] = 2018
data2017 = data2017[["National Provider Identifier", "Average Patient Payment Amount"]]
data2017["Year"] = 2017
data2016 = data2016[["National Provider Identifier", "Average Patient Payment Amount"]]
data2016["Year"] = 2016
data2015 = data2015[["National Provider Identifier", "Average Patient Payment Amount"]]
data2015["Year"] = 2015
data2014 = data2014[["National Provider Identifier", "Average Patient Payment Amount"]]
data2014["Year"] = 2014
data2013 = data2013[["National Provider Identifier", "Average Patient Payment Amount"]]
data2013["Year"] = 2013
data2012 = data2012[["National Provider Identifier", "Average Patient Payment Amount"]]
data2012["Year"] = 2012
```

```python
data2018
```

```python
final_df = pd.concat([data2018, data2017, data2016, data2015, data2014, data2013, data2012])
final_df = final_df.set_index("National Provider Identifier")
```

```python
final_df
```

```python
final_df_pivot = final_df.pivot(columns = "Year", values = "Average Patient Payment Amount")
final_df_pivot.columns = ["Average Patient Payment in " + str(i) for i in final_df_pivot.columns]
```

```python
final_df_pivot.to_csv("Processed Datasets/average_patient_payment_by_year.csv")
```
